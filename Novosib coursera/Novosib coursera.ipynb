```r
advanced_summary <- function(df, basic = 'true', sdo = 'false', mad = 'false', q_q = 'false', razmah = 'false') {
    for(i in names(df)) {
        print(i)
        cat('\n')
        if (basic == 'true') {
            print(summary(df[[i]]))
            cat('\n')
        }
        if (!is.factor(df[[i]])) {
            if (sdo == 'true') {
                cat('SDO:', sd(df[[i]]), "\n\n") 
            }
            if (mad == 'true') {
                cat('MAD:', mad(df[[i]]), "\n\n") 
            }
            if (q_q == 'true') {
                cat('Q-Q:', IQR(df[[i]]), "\n\n") 
            }
            if (razmah == 'true') {
                cat('Razmah:', max(df[[i]]) - min(df[[i]]), "\n\n") 
            }
            cat('-------------------------------------------------------------', '\n\n\n')
        }
    }
}
    
```

```r
advanced_summary(data, sdo = 'true', mad = 'true', q_q = 'true', razmah = 'true')
```
